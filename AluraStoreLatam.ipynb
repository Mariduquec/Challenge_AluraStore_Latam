```python
# --- Utilitarias ---
def find_column(df, candidates):
    cols = list(df.columns)
    # match exact (case-insensitive)
    for cand in candidates:
        for c in cols:
            if c.strip().lower() == cand.strip().lower():
                return c
    # match contains
    for cand in candidates:
        for c in cols:
            if cand.strip().lower() in c.strip().lower():
                return c
    return None

def clean_numeric_series(series):
    """
    Convierte una serie con valores numéricos en texto a float.
    - Maneja símbolos de moneda, espacios y caracteres no numéricos.
    - Interpreta separadores de miles (.) y decimales (,) o viceversa.
    - Devuelve NaN en valores vacíos o inválidos.
    """
    def parse_value(x):
        if pd.isna(x):
            return np.nan

        s = str(x).strip()
        if s == '':
            return np.nan

        # Eliminar cualquier caracter no numérico válido
        s = re.sub(r'[^0-9\-,\.]', '', s)
        if s in ['', '-', ',']:
            return np.nan

        # Caso con punto y coma → '.' como miles y ',' como decimal
        if s.count('.') > 0 and s.count(',') > 0:
            s = s.replace('.', '').replace(',', '.')
        # Solo coma → coma como decimal
        elif s.count(',') > 0 and s.count('.') == 0:
            s = s.replace(',', '.')
        # Solo puntos y más de uno → primeros como miles, último como decimal
        elif s.count('.') > 1 and s.count(',') == 0:
            parts = s.split('.')
            s = ''.join(parts[:-1]) + '.' + parts[-1]

        # Intentar conversión a float
        try:
            return float(s)
        except:
            return np.nan

    return series.apply(parse_value)
```

#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
# --- Cálculo de ingreso por tienda ---
rows = []
for name, link in urls.items():
    try:
        df = pd.read_csv(link)
    except Exception as e:
        print(f"[ERROR] No se pudo cargar {name} desde {link}: {e}")
        rows.append({'tienda': name, 'n_transacciones': np.nan, 'total_revenue': np.nan})
        continue

    # Detectar columnas comunes
    price_col = find_column(df, ['valor', 'precio', 'price', 'total', 'monto', 'amount'])
    qty_col = find_column(df, ['cantidad', 'quantity', 'qty', 'qtd', 'units'])

    # Mostrar columnas si no encuentra precio
    if price_col is None:
        print(f"[AVISO] No se detectó columna de precio en {name}. Columnas disponibles: {list(df.columns)}")
    else:
        try:
            df[price_col] = clean_numeric_series(df[price_col])
        except Exception as e:
            print(f"[ERROR] Al limpiar precio en {name}: {e}")

    # Cantidad: si no existe, asumimos 1 por transacción
    if qty_col is None:
        df['_cantidad_temp'] = 1
        qty_col = '_cantidad_temp'
    else:
        try:
            df[qty_col] = clean_numeric_series(df[qty_col]).fillna(1)
        except Exception as e:
            print(f"[ERROR] Al limpiar cantidad en {name}: {e}")
            df[qty_col] = df.get(qty_col, pd.Series([1]*len(df)))

    # Calcular ingresos
    if price_col is not None:
        df['_revenue'] = df[price_col] * df[qty_col]
        total_revenue = df['_revenue'].sum(skipna=True)
    else:
        total_revenue = np.nan

    rows.append({
        'tienda': name,
        'n_transacciones': int(df.shape[0]),
        'total_revenue': float(total_revenue) if not pd.isna(total_revenue) else np.nan
    })

# Resultado final
result = pd.DataFrame(rows).sort_values('tienda').reset_index(drop=True)
result['total_revenue_formatted'] = result['total_revenue'].apply(lambda x: f"{x:,.2f}" if pd.notna(x) else "N/A")

print("\nIngreso total por tienda:")
display(result[['tienda', 'n_transacciones', 'total_revenue_formatted']])
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
# Column name candidates
cat_candidates = ['categoria', 'category', 'cat', 'categoria_producto', 'tipo', 'category_name']
qty_candidates = ['cantidad', 'quantity', 'qty', 'qtd', 'units', 'unidades']

# Diccionarios para resultados
ventas_por_categoria_dict = {}
filas_resumen = []

for tienda_name, link in urls.items():
    try:
        df = pd.read_csv(link)
    except Exception as e:
        print(f"[ERROR] No se pudo cargar {tienda_name}: {e}")
        continue

    # detectar columnas
    cat_col = find_column(df, cat_candidates)
    qty_col = find_column(df, qty_candidates)

    # si no hay columna de categoría, avisar y saltar
    if cat_col is None:
        print(f"[AVISO] No se detectó columna de categoría en {tienda_name}. Columnas: {list(df.columns)}")
        continue

    # normalizar cantidad (si existe) o asumir 1 por fila
    if qty_col is None:
        df['_cant_temp'] = 1
        qty_col_use = '_cant_temp'
    else:
        df[qty_col] = clean_numeric_series(df[qty_col]).fillna(1)
        qty_col_use = qty_col

    # Agrupar por categoría y sumar cantidad
    cat_counts = (df
                  .groupby(cat_col)[qty_col_use]
                  .sum()
                  .rename('units_sold')
                  .sort_values(ascending=False))

    ventas_por_categoria_dict[tienda_name] = cat_counts

    # preparar filas resumen (tabla consolidada)
    tmp = cat_counts.reset_index().rename(columns={cat_col: 'categoria'})
    tmp['tienda'] = tienda_name
    filas_resumen.append(tmp[['tienda', 'categoria', 'units_sold']])

# DataFrame consolidado con todas las tiendas
if filas_resumen:
    ventas_por_categoria = pd.concat(filas_resumen, ignore_index=True)
    ventas_por_categoria = ventas_por_categoria.sort_values(['tienda', 'units_sold'], ascending=[True, False])
    print("Ventas por categoría (consolidado):")
    display(ventas_por_categoria)
else:
    ventas_por_categoria = pd.DataFrame(columns=['tienda','categoria','units_sold'])
    print("No se generó consolidado (no se detectaron categorías).")

# Mostrar top 5 categorías por tienda y graficar
for tienda_name, series in ventas_por_categoria_dict.items():
    print(f"\nTop categorías - {tienda_name}:")
    display(series.head(10).reset_index().rename(columns={series.index.name or cat_col: 'categoria', 0: 'units_sold'}))
    # Plot top 5
    top = series.head(5)
    if len(top) > 0:
        plt.figure(figsize=(8,4))
        top.plot(kind='bar')
        plt.title(f"Top 5 categorías vendidas - {tienda_name}")
        plt.ylabel("Unidades vendidas")
        plt.xlabel("Categoría")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    else:
        print(f"No hay categorías para graficar en {tienda_name}.")

# # Guardar resultado consolidado (opcional)
# ventas_por_categoria.to_csv('ventas_por_categoria_consolidado.csv', index=False)
# print("\nSe guardó 'ventas_por_categoria_consolidado.csv' en el directorio de trabajo (si se generó).")
# ===== fin celda =====
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
# --- Función para obtener productos más y menos vendidos ---
def productos_vendidos(df, tienda):
    # Contar cuántas veces aparece cada producto (no hay columna de cantidad)
    ventas = df['Producto'].value_counts()

    # Top 5 más vendidos
    top5 = ventas.head(5)
    # Bottom 5 menos vendidos
    bottom5 = ventas.tail(5)

    print(f"\n===== {tienda} =====")
    print("Top 5 productos más vendidos:")
    print(top5)
    print("\nBottom 5 productos menos vendidos:")
    print(bottom5)

    # Gráficos
    plt.figure(figsize=(8,4))
    top5.plot(kind='bar', color='green', title=f"{tienda} - Top 5 más vendidos")
    plt.ylabel("Número de ventas")
    plt.xticks(rotation=45, ha='right')
    plt.show()

    plt.figure(figsize=(8,4))
    bottom5.plot(kind='bar', color='red', title=f"{tienda} - Bottom 5 menos vendidos")
    plt.ylabel("Número de ventas")
    plt.xticks(rotation=45, ha='right')
    plt.show()

# --- Aplicar a cada tienda ---
for nombre, df in tiendas.items():
    productos_vendidos(df, nombre)
```
